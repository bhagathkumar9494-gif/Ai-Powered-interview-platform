**Data Cleaning**

```python
#df.isnull() or df.isna() handles the missing values
#dropna() removes the rows with missing values
#fillna(val) fill the missing values with some value called val
#duplicated() identifies the duplicates
#drop_duplicates() removes duplicate rows->option=(keep='first')=>keeps first occureance==>(keep='second')=>keeps the second occurance
```

```python
catogorical_columns = df.select_dtypes(include=['object','category']).columns.tolist()
#select_dtypes is used to select all the columns in the with the specified data types => if we need only columns names do .columns and to to covert the names to list instead of object do .tolist()
```

```python
df['Venue'].unique().tolist()
```

```python
 venue_map = {
     "M.Chinnaswamy Stadium": "M Chinnaswamy Stadium",
     "Sardar Patel Stadium": "Narendra Modi Stadium",
     "Punjab Cricket Association IS Bindra Stadium": "Punjab Cricket Association Stadium",
     "Feroz Shah Kotla": "Arun Jaitley Stadium",
     "Zayed Cricket Stadium": "Sheikh Zayed Stadium",
     "De Beers Diamond Oval": "Kimberley Oval",
     "Rajiv Gandhi International Stadium": "Uppal Stadium",
     "Dr Y.S. Rajasekhara Reddy ACA-VDCA Cricket Stadium": "ACA-VDCA Stadium",
     "Shaheed Veer Narayan Singh International Stadium": "Raipur International Cricket Stadium",
     "JSCA International Stadium Complex": "Ranchi Stadium",
     "Barsapara Cricket Stadium": "Assam Cricket Association Stadium"
 }

 df['Venue'] = df['Venue'].replace(venue_map) #replace is used to replace each row value of specified column with new value
```

```python
teamNameChange = {
     'Kings XI Punjab': 'Punjab Kings',
     'Delhi Daredevils': 'Delhi Capitals',
}
df['Bat First'] = df['Bat First'].replace(teamNameChange)
df['Bat Second'] = df['Bat Second'].replace(teamNameChange)
df['Winner'].unique().tolist()
```

```python
df['Batter']=df['Batter'].str.strip().str.title()
df['Non Striker'] = df['Non Striker'].str.strip().str.title()
df["Bowler"] = df["Bowler"].str.strip().str.title()
df["Player Out"] = df["Player Out"].str.strip().str.title()
```

```python
df.dropna()#used to remove rows with missing values
```

```python
df.isna().sum() # used to know the number of missing values rows in each columns isna() gives boolean based on that 1,0 sum is calculated
```

```python
df['Method'].fillna('Not Out', inplace=True) # fillna() is used to fill the missing row values with some other values that we have provided and instead of assigning to column the current changed row the function has one more argument called inplace whose value by default is false in order too have a direct changes in the rows of particular coulmn make inplace=True
df['Player Out'].fillna('None', inplace=True)
df['Player Out Runs'].fillna(0, inplace=True)
df['Player Out Balls Faced'].fillna(0, inplace=True)
df['Runs to Get'].fillna(0, inplace=True)
df.isnull().sum()#cross verify if every coulmn has no missing values
```
